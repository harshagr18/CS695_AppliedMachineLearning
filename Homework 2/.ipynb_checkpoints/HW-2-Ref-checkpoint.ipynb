```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
val = np.seterr(all='ignore')
```

```python
df = pd.read_csv('data2.txt', sep=",", header=None)
msk = np.random.rand(len(df)) < 0.8
df.columns = ["x-axis", "y-axis"]
```

```python
df.plot(kind='scatter',x='x-axis',y='y-axis',color='red')
plt.show()
```

```python
#Splitting the dataset into 80% training data and 20% test data.
```

```python
train = df[msk]
test = df[~msk]
```

```python
x_train = train['x-axis'].values
y_train = train['y-axis'].values
m = len(train)
x_bias_train = np.ones((m,1))
x_train = np.reshape(x_train,(m,1))
x_train = np.append(x_bias_train,x_train,axis=1)
x_transpose_train = np.transpose(x_train)
x_transpose_dot_x_train = x_transpose_train.dot(x_train)
temp_1_train = np.linalg.inv(x_transpose_dot_x_train)
temp_2_train=x_transpose_train.dot(y_train)
theta_train =temp_1_train.dot(temp_2_train)
print("Value of the coefficients is: ",theta_train)
```

```python
y_train = theta_train[0] + theta_train[1]*x_train
train.plot(kind='scatter',x='x-axis',y='y-axis',color='red')
plt.plot(x_train,y_train,'-',color='blue')
plt.show()
```

```python
m = len(train)
x_train = np.reshape(train['x-axis'].values,(m,1))
y_train = np.reshape(train['y-axis'].values,(m,1))
```

```python
def cal_cost(theta,X,y):
    cost= np.sum(np.square(((np.dot(X, theta)) - y))) / (2 * (len(X)))  
    return cost
```

```python
def gradient_descent(x,y,theta,learning_rate=0.01,iterations=2000):
    m = len(y)
    cost_history = np.zeros(iterations)
    theta_history = np.zeros((iterations,2))
    for it in range(iterations):
        prediction = np.dot(x,theta)
        theta = theta -(1/m)*learning_rate*( x.T.dot((prediction - y)))
        theta_history[it,:] = theta.T
        cost_history[it]  = cal_cost(theta,x,y)
    return theta, cost_history, theta_history
```

```python
lr =0.02
n_iter = 2000
theta_train = np.random.randn(2,1)
X_b_train = np.c_[np.ones((len(train),1)),x_train]
theta_train,cost_history,theta_history = gradient_descent(X_b_train,y_train,theta_train,lr,n_iter)
print('Theta0:          {:0.3f},\nTheta1:          {:0.3f}'.format(theta_train[0][0],theta_train[1][0]))
print('Final cost/MSE:  {:0.3f}'.format(cost_history[-1]))
print('-------------------------')
for it in range(n_iter):
    print("Iterations: ",it," Cost Function Value: ",round(cost_history[it],3))
```

```python
fig,ax = plt.subplots(figsize=(6,4))

ax.set_ylabel('The value of the cost function J')
ax.set_xlabel('Total number of iterations')
_=ax.plot(range(n_iter),cost_history,'r-')
```

```python
#Now we calculate the accuracy of the testing set w.r.t. to the Thera values found while using the Training Set.
```

```python
m = len(test)
x_test = np.reshape(test['x-axis'].values,(m,1))
y_test = np.reshape(test['y-axis'].values,(m,1))

lr =0.02
n_iter = 2000
theta_test = theta_train
X_b_test = np.c_[np.ones((len(test),1)),x_test]
theta_test,cost_history,theta_history = gradient_descent(X_b_test,y_test,theta_test,lr,n_iter)
print('Final cost/MSE:  {:0.3f}'.format(cost_history[-1]))
print('-------------------------')
for it in range(n_iter):
    print("Iterations: ",it," Cost Function Value: ",round(cost_history[it],3))
```

```python
#Hence, here we can see that the cost function for the testing set is significantly less as compared to the
#cost function for the training set.
```

```python
#Stochastic Gradient Descent
```

```python
m = len(train)
x_train = np.reshape(train['x-axis'].values,(m,1))
y_train = np.reshape(train['y-axis'].values,(m,1))
```

```python
def stocashtic_gradient_descent(X,y,theta,learning_rate=0.01,iterations=10):
    m = len(y)
    cost_history = np.zeros(iterations) 
    for it in range(iterations):
        cost =0.0
        for i in range(m):
            rand_ind = np.random.randint(0,m)
            X_i = X[rand_ind,:].reshape(1,X.shape[1])
            y_i = y[rand_ind].reshape(1,1)
            prediction = np.dot(X_i,theta)

            theta = theta -(1/m)*learning_rate*( X_i.T.dot((prediction - y_i)))
            cost += cal_cost(theta,X_i,y_i)
        cost_history[it]  = cost
        
    return theta, cost_history
```

```python
lr =0.01
n_iter = 2500
theta_train = np.random.randn(2,1)
X_b_train = np.c_[np.ones((len(train),1)),x_train]
theta_train,cost_history = stocashtic_gradient_descent(X_b_train,y_train,theta_train,lr,n_iter)
print('Theta0:          {:0.3f},\nTheta1:          {:0.3f}'.format(theta_train[0][0],theta_train[1][0]))
print('Final cost/MSE:  {:0.3f}'.format(cost_history[-1]))
print('-------------------------')
for it in range(n_iter):
    print("Iterations: ",it," Cost Function Value: ",round(cost_history[it],3))
```

```python
fig,ax = plt.subplots(figsize=(12,8))
ax.set_ylabel('J(Cost Function)')
ax.set_xlabel('Iterations')
_=ax.plot(range(n_iter),cost_history,'r')
```

```python
m = len(test)
x_test = np.reshape(test['x-axis'].values,(m,1))
y_test = np.reshape(test['y-axis'].values,(m,1))

lr =0.01
n_iter = 2500
theta_test = theta_train
X_b_test = np.c_[np.ones((len(test),1)),x_test]
theta_test,cost_history = stocashtic_gradient_descent(X_b_test,y_test,theta_test,lr,n_iter)
print('Final cost/MSE:  {:0.3f}'.format(cost_history[-1]))
print('-------------------------')
for it in range(n_iter):
    print("Iterations: ",it," Cost Function Value: ",round(cost_history[it],3))
```

```python
#Hence, here we can see that the cost function for the testing set is significantly less as compared to the
#cost function for the training set.
#Hence as the test data set error is significantly low, we can state that our model does not Overfit.
```
