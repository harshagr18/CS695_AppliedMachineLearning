```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
from sklearn.metrics import mean_squared_error
noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.5)
y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
```

```python
a=[]
b=[]
for i in range(len(x)):
    a.append(x[i][0])
    b.append(y[i][0])
```

```python
fa = np.polyfit(a,b,4)
fx = np.poly1d(fa)

plt.plot(a,b, 'ro')
plt.plot(a,fx(a), 'bo')
plt.show()
```

```python
l=[]
for i in [1,2,3,4,5,6,7,8]:
    fa = np.polyfit(a,b,i)
    fx = np.poly1d(fa)
    l.append(mean_squared_error(b,fx(a)))
plt.plot([1,2,3,4,5,6,7,8],l,'g')
```

From the observation, we can say that the elbow point is obtained when m = 3 i.e. a polynomial of order 3 is able to understand the data.  - Note: After m=3, the value of error plateaus

```python
for i in [200, 300, 400, 600, 800, 1000 ]:
    noise_scale = i
    number_of_samples = 50
    x = 25*(np.random.rand(number_of_samples, 1) - 0.5)
    y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
    a=[]
    b=[]
    for i in range(len(x)):
        a.append(x[i][0])
        b.append(y[i][0])
    fa = np.polyfit(a,b,3)
    fx = np.poly1d(fa)

    plt.plot(a,b, 'ro')
    plt.plot(a,fx(a), 'bo')
    plt.show()
```

We can observe that noise is directly proportional to the error.

```python
for i in [40, 30, 20, 10 ]:
    noise_scale = 100
    number_of_samples = i
    x = 25*(np.random.rand(number_of_samples, 1) - 0.5)
    y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
    a=[]
    b=[]
    for i in range(len(x)):
        a.append(x[i][0])
        b.append(y[i][0])
    fa = np.polyfit(a,b,3)
    fx = np.poly1d(fa)

    plt.plot(a,b, 'ro')
    plt.plot(a,fx(a), 'bo')
    plt.show()
```

As the data sample decreases, it becomes more difficult for the polynomial to find patterns in the underlying data.
