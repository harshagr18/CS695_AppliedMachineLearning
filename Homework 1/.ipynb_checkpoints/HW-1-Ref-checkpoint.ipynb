```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid') 
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python

#The polynomial is y = 7 * x + 15 * x**2 + 1 * x**3

noise_scale = 100 
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
y = 7 * x + 15 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1) 
plt.plot(x,y,'ro')
```

```python


#The value of m selected is 4.


x = x.ravel()
y = y.ravel()
z = np.polyfit(x, y, 4)
p = np.poly1d(z)
plt.plot(x, y, 'ro', label = 'Noisy Data')
plt.plot(x,p(x),'bo',label = 'Polynomial')
plt.legend()
plt.show()
```

```python
x = x.ravel()
y = y.ravel()
m = []
for i in [1, 2, 3, 4, 5, 6, 7, 8]:
    z = np.polyfit(x, y, i)
    p = np.poly1d(z)
    m.append(((sum((y-p(x))**2)))/number_of_samples)
    plt.title(f'M = {i}')
    plt.plot(x, y, 'ro', label = 'Noisy Data')
    plt.plot(x,p(x),'bo',label = 'Polynomial')
    plt.legend()
    plt.show()
print("MSE: ",m)
plt.plot([1, 2, 3, 4, 5, 6, 7, 8],m)
plt.xlabel('M')
plt.ylabel('MSE')
plt.show()
```

```python
#According to the Plotted Graph, the best value of m is 3.
```

```python
for noise_scale in [150, 200, 400, 600, 1000]: 
    number_of_samples = 50
    x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
    y = 7 * x + 15 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1) 
    x = x.ravel()
    y = y.ravel()
    z = np.polyfit(x, y, 3)
    p = np.poly1d(z)
    plt.title(f'noise_scale = {noise_scale}')
    plt.plot(x, y, 'ro', label = 'Noisy Data')
    plt.plot(x,p(x),'bo',label = 'Polynomial')
    plt.legend()
    plt.show()
```

```python
#As the noise scale increases, the noisy Data is more scattered, and the noisy points start moving away
#from our polynomial.
```

```python
import matplotlib.pyplot as plt
import matplotlib

plt.style.use('seaborn-whitegrid')
import numpy as np

plt.figure(0)
plt.title(f'Erros vs Sample_Space_Size')

for i in [40, 30, 20, 10]:
  plt.figure(i)
  noise_scale = 100
  x = 25*(np.random.rand(i, 1) - 0.8)
  y = 7 * x + 15 * x**2 + 1 * x**3 + noise_scale*np.random.randn(i, 1)
  plt.plot(x,y,'ro', label='Input')

  x = x.ravel()
  y = y.ravel()
  t = np.linspace(-20,5)

  trend = np.polyfit(x,y,3)
  plt.plot(t, np.polyval(trend, t), label='Given Polynomial')
  plt.title(f'Sample_Size={i}')
  plt.legend()
  mse = ((y - np.polyval(trend,x))**2).mean(axis=0)
  plt.figure(0)
  plt.plot(i, mse, 'ro')
```

```python
#Here we can see that as the number of samples increase, the error increases.
#Hence, here it is observed that the Increase in the Sample Space does not lead to a better solution.
```
