```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.5)
y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
```

```python
# Let's take m as 2

x = x.ravel()
y = y.ravel()


func = np.polyfit(x, y, 2)
q = np.poly1d(func)
plt.plot(x, y, 'ro', label = 'Data')
plt.plot(x,q(x),'bo',label = 'PolyFit Function')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

MSE = []

for i in [1, 2, 3, 4, 5, 6, 7, 8]:
    func = np.polyfit(x, y, i)
    q = np.poly1d(func)
    MSE.append(mean_squared_error(y,q(x)))

plt.plot([1, 2, 3, 4, 5, 6, 7, 8],MSE)
plt.xlabel('Order(M)')
plt.ylabel('MSE')
plt.show()
```

### From the plot it is obvious, that the best value of M is 3.



```python
MSE = []

for noise_scale in [200, 300, 400, 600, 800, 1000]: 
    number_of_samples = 50
    x = 25*(np.random.rand(number_of_samples, 1) - 0.5)
    y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1) 
    x = x.ravel()
    y = y.ravel()
    func = np.polyfit(x, y, 3)
    q = np.poly1d(func)
    MSE.append(mean_squared_error(y,q(x)))
    plt.title(str('Plot for noise scale = '+str(noise_scale)))
    plt.plot(x, y, 'ro', label = 'Data')
    plt.plot(x,q(x),'bo',label = 'PolyFit Function')
    plt.legend()
    plt.show()
    
    
plt.plot([200, 300, 400, 600, 800, 1000],MSE)
plt.xlabel('Noise Scale')
plt.ylabel('MSE')
plt.show()
```

### As the noise increases, variance of the dataset increases and the Mean Squared Error of the polynomial on the dataset will also increase.

```python
MSE = []

for number_of_samples in [40, 30, 20, 10]: 
    noise_scale = 100
    x = 25*(np.random.rand(number_of_samples, 1) - 0.5)
    y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1) 
    x = x.ravel()
    y = y.ravel()
    func = np.polyfit(x, y, 3)
    q = np.poly1d(func)
    MSE.append(mean_squared_error(y,q(x)))
    plt.title(str('Plot for number of samples = '+str(number_of_samples)))
    plt.plot(x, y, 'ro', label = 'Data')
    plt.plot(x,q(x),'bo',label = 'PolyFit Function')
    plt.legend()
    plt.show()
    
    
plt.plot([40, 30, 20, 10],MSE)
plt.xlabel('Number of Samples')
plt.ylabel('MSE')
plt.show()
```

### With a low number of samples, the polynomial is unable to figure out the actual polynomial required to fit the dataset. Hence a large number of data points are required for the polynomial
