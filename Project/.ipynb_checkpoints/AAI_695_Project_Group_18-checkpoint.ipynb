# Introduction
We are interested in exploring the ability to predict a borrower’s likelihood to pay back a loan based on available information at the origination of a loan.

This dataset provides us with the loan information from Lending Club, one of the world’s largest peer-to-peer lending platforms ( https://www.kaggle.com/wordsforthewise/lending-club ).The dataset includes important information about a large amount of specific loans. We will use the features, some describing the loan(e.g., the monthly payment and interest rate) while others are about borrowers(e.g., borrowers’ job title and annual income), to **predict whether this loan will be finally paid off**. It is a **binary-classification** task.

# Our Goal
- We will use exploratory data analysis, including statistical analysis and visualization analysis, to extract features that significantly influence the loan status: whether this loan would be paid off. It is a large dataset with millions of loan data and more than 20 features. What’s more, not all features are numerical. Some missing values exists. We will show how we deal with this challenging dataset and make data-driven conclusions.
- We will implement several models for our classification, visualize and explain each followed by comparing their methods and results. Our plan is to include:
 * LDA and Maximum Likelihood Estimation
 * Logistic regression and RandomForest
 * XGBoost
- The amount of loans being paid off is far more than the number of loans that were charged off, so we meet data imbalance problem here. We will try to solve this problem from both the insight of data processing and machine learning model development.
 * Data processing insight: Develop sampling method. We would try both oversampling and down-sampling.
 * Model insight: Adjust the weight of data from different categories in the objective function for model to optimize.
We will make final model evaluation by F1-score, PR or AUC curve instead of accuracy, which is proved meaningless while data imbalance problem exists. Given those scores and our understanding of the approaches, we will attempt to relate their performances to our expectations.

### Preprocessing

```python
# Since the data has columns consisting of up to 90% NaN values, we filter all columns containing more than 20% NaN

percent_df = df.isna().mean()
x = percent_df[percent_df*100>20]
df.drop(x.index,axis=1,inplace=True)

df.head()
```

```python
plt.figure(figsize=(18,4),dpi=300)
sns.barplot(y=((df.isnull().sum()/len(df))*100).sort_values(), x=((df.isnull().sum()/len(df))*100).sort_values().index, palette='viridis') 
plt.xticks(rotation=90)
```

```python
# Now dropping some of the unnecessary columns

delete_cols = ['id','home_ownership','verification_status','issue_d','url','title','zip_code',"addr_state",'earliest_cr_line',
'last_pymnt_d','last_pymnt_amnt','last_credit_pull_d']

set(x.index).intersection(set(delete_cols))

df.drop(delete_cols,axis=1,inplace=True)
df.head()
```

```python
# Now fixing the target column, by combining categories and considering binary classification using the 2 major categories

df['loan_status'] = df['loan_status'].replace(['Does not meet the credit policy. Status:Fully Paid'],'Fully Paid')
df['loan_status'] = df['loan_status'].replace(['Does not meet the credit policy. Status:Charged Off'],'Charged Off')
filt_df = df[(df['loan_status'] == 'Fully Paid') | (df['loan_status'] == 'Charged Off')]

filt_df['loan_status'].value_counts()
```

```python
# Dropping 2 extra unnecessary columns

filt_df.drop('emp_title',axis=1,inplace=True)
filt_df.drop('emp_length',axis=1,inplace=True)
```

```python
# Building lists of categorical and numerical columns

cat_col = list(filt_df.select_dtypes('object'))
num_col = list(filt_df.select_dtypes('float64'))

# Replacing numerical column NaNs by their mean

x = filt_df[num_col].isna().mean()
nan_columns= x[x>0].index
filt_df['num_sats'].fillna(filt_df['num_sats'].mean()).head()
for col in nan_columns:
    filt_df[col] = filt_df[col].fillna(filt_df[col].mean())
filt_df[nan_columns].isna().sum()
```

As evident, NaNs from numerical columns have become 0

```python
# Now replacing NaNs from categorical columns

for col in cat_col:
    filt_df[col] = filt_df[col].fillna(filt_df[col].mode())
    
filt_df[cat_col].isna().sum()
```

As evident, NaNs from categorical columns have become 0

```python
# Finally we convert categorical values to numerical

from tqdm import tqdm
for col in tqdm(cat_col):
    filt_df[col] = filt_df[col].replace(filt_df[col].unique(),list(range(len(filt_df[col].unique()))))
```

## Over/Undersampling
Apply oversampling and undersampling and train a logistic regression model to observe impact of balancing the samples.

Synthetic Minority Oversampling Technique (SMOTE) is a statistical technique for increasing the number of cases in your dataset in a balanced way

```python
# Oversampling

from imblearn.over_sampling import SMOTE
smt = SMOTE()
X_smote, y_smote = smt.fit_resample(X_train, y_train)

sns.countplot(y_train)
```

```python
sns.countplot(y_smote)
```

```python
# build the model
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression(random_state= 42)

# fit the model on training data
lr_model.fit(X_smote,y_smote)

# make prediction on test data
y_pred_smote = lr_model.predict(X_test)

f1_results.append({
    'Name':'Oversample-LR', 
    'f1-pos':sklearn.metrics.f1_score(y_test, y_pred_smote), 
    'f1-neg':sklearn.metrics.f1_score(1-y_test, 1-y_pred_smote)})
print(classification_report(y_test,y_pred_smote))
```

# Statistical Analysis

## Sample Z test (Column wise)
- we will implement 2 sample z-test where one variable will be categorical with two categories and the other variable will be continuous to apply the z-test.


Null Hypothesis: There is no significant difference between the mean loan amount of paid and charged off loans.

Alternate Hypothesis: There is a significant difference between the mean loan amount of paid and charged off loans.


NOTE - we can also use t-test but it is used to compare the mean of two given samples like the Z-test. However, It is implemented when the sample size is less than 30. It assumes a normal distribution of the sample. It can also be one-sample or two-sample.

- **Loan Status** - '0' - Fully Paid
- **Loan Status** - '1' - Charged Off

```python
df = filt_df.copy()
list_float = filt_df.select_dtypes(exclude=['object']).columns
list_object = ['term', 'grade', 'sub_grade', 'pymnt_plan', 'purpose']

for col in list_object:
    finalDF[col] = finalDF[col].astype('object')
```

# Loan Amount

```python
from numpy import sqrt
from scipy.stats import norm
full_paid_mean=df.loc[df['loan_status']==0,'loan_amnt'].mean()
charged_off_mean=df.loc[df['loan_status']==1,'loan_amnt'].mean()
full_paid_std=df.loc[df['loan_status']==0,'loan_amnt'].std()
charged_off_std=df.loc[df['loan_status']==1,'loan_amnt'].std()
no_of_full_paid=df.loc[df['loan_status']==0,'loan_amnt'].count()
no_of_charged_off=df.loc[df['loan_status']==1,'loan_amnt'].count()

def twoSampZ(X1, X2, mudiff, sd1, sd2, n1, n2):
    pooledSE = sqrt(sd1**2/n1 + sd2**2/n2)
    z = ((X1 - X2) - mudiff)/pooledSE
    pval = 2*(1 - norm.cdf(abs(z)))
    return round(z,3), pval
z,p= twoSampZ(full_paid_mean,charged_off_mean,0,full_paid_std,charged_off_std,no_of_full_paid,no_of_charged_off)
print('Z', z)
print('p', p)

if p<0.05:
    print("We reject null hypothesis and accept Alternate Hypothesis")
    print('Alternate Hypothesis: There is a significant difference between the mean loan amount of paid and charged off loans.')
else:
    print("We accept null hypothesis")    
```

# CONCLUSIONS

We are able to make a number of observations from this project:
- Understanding the data is key - When I first started and used the full set of data from the .csv file, I was able to predict very well both defaulting and paying off of loans. As I worked to understand the column data better, I discovered that many of the columns had information related to a loan nearing default or in default (example. How many months since the last payment). Removal of those columns made the problem more realistic and more difficult.
- Preprocessing the data is very important - Dealing with missing values in an appropriate manner enables the use of a large dataset. Just removing the offending samples can inadvertently lead to biases in the data so I decided to try to fill in the values with minimum-bias values.
- Sophistication doesn't necessarily mean better – I tried a cross-sample from a very straightforward PCA method to Logistic Regression with very similar results. That's not to say there's not an appropriate time to apply more complicated models but these results would imply there's not always a strong correspondence between the available sample data and the desired output regardless of how you slice it.
